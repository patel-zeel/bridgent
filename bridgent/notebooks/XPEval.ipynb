```python
def load_xp(fnm):
    values = []
    for i, line in enumerate(open(fnm)):
        if (i + 1) % 3 == 0:
            values.append(eval(line)[0])
    return values
```

```python
n_window = 50
plt.plot(np.convolve(load_xp('./xpnt/deepres_do_0_6_l2_0_05.txt'), np.ones(n_window)/n_window)[300:-n_window], alpha=0.6)
plt.plot(np.convolve(load_xp('./xpnt/deepres_do_0_7_l2_0_05.txt'), np.ones(n_window)/n_window)[300:-n_window], alpha=0.6);
```

`deepres_do_0_6_l2_0_05.txt` is better
